### 4.5 DNN

```python
import tensorflow as tf
from tensorflow import keras
from keras import models, layers
from scikeras.wrappers import KerasClassifier
from scikeras.wrappers import KerasRegressor
from sklearn.model_selection import KFold
```

```python
def finalScoreAnn(annMetrics, expNum):
    print('Accuracy',np.mean(annMetrics[0]))
    print('precision',np.mean(annMetrics[1]))
    print('recall',np.mean(annMetrics[2]))
    print('f1_score ',np.mean(annMetrics[3]))

    experimentResults.append([expNum , np.mean(annMetrics[0]), np.mean(annMetrics[1]),np.mean(annMetrics[2]),np.mean(annMetrics[3])])
    
```

```python
def metricStore(y_test, annPred,annMetrics):
    accuracy = accuracy_score(y_test, annPred.round())
    recall = recall_score(y_test, annPred.round(), average='macro')
    precision = precision_score(y_test, annPred.round(), average='macro')
    F1_macro = f1_score(y_test, annPred.round(), average='macro')
    #store results        
    annMetrics[0].append(accuracy)
    annMetrics[1].append(precision)
    annMetrics[2].append(recall)
    annMetrics[3].append(F1_macro)
    print(annMetrics)
    
```

```python
def kfoldDNN(feature, target, expNum):
    kf=KFold(n_splits=10)
    dnnMetrics = [[],[],[],[]]
    for train_index, test_index in kf.split(feature, target):
         #test train split according to K fold
         X_train, X_test = feature.iloc[train_index], feature.iloc[test_index]
         y_train, y_test = target.iloc[train_index], target.iloc[test_index]
         #ann model
         DNNmodel = keras.Sequential([
            layers.Dense(16, activation="relu", input_dim=(X_train.shape[1])),
            layers.Dense(16, activation="relu"),
            layers.Dense(1, activation="sigmoid")
            ])
         # Compile model
         DNNmodel.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=["accuracy", "precision","recall","f1_score"] )
         DNNHistory=DNNmodel.fit(X_train, y_train, epochs=20, batch_size=512, validation_data=(X_test , y_test))
         
         DNNPred = DNNmodel.predict(X_test)
         metricStore(y_test, DNNPred,dnnMetrics)

    print(finalScoreAnn(dnnMetrics, expNum))
         


```
