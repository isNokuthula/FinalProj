```python
def meanDnn2(dnnMetrics):
    print('Accuracy',np.mean(dnnMetrics[0]))
    print('precision',np.mean(dnnMetrics[1]))
    print('recall',np.mean(dnnMetrics[2]))
    f1Alt = (2 * (np.mean(dnnMetrics[1]) * np.mean(dnnMetrics[2])) / (np.mean(dnnMetrics[1]) + np.mean(dnnMetrics[2])))
    print('f1_score 2',f1Alt)
    print('f1_score',np.mean(dnnMetrics[3]))
```

```python
dnnMetrics = [[],[],[],[],[]]
def kfoldDNN(feature, target):
    kf=KFold(n_splits=2)
    for train_index, test_index in kf.split(feature, target):
         #print(train_index)
         #print(test_index)
        #X.iloc[:3993,:]
         X_train, X_test = feature.iloc[train_index], feature.iloc[test_index]
         y_train, y_test = target.iloc[train_index], target.iloc[test_index]
         # Compile model
         model.compile(loss='binary_crossentropy', optimizer='adam', metrics=["accuracy", "precision","recall","f1_score"], )
         model.fit(X_train, y_train, epochs=5, batch_size=512, validation_data=(X_test , y_test))
         #history.history['f1_score']
         dnnMetrics[0].append(history.history['accuracy'])
         dnnMetrics[1].append(history.history['precision'])
         dnnMetrics[2].append(history.history['recall'])
         dnnMetrics[3].append(history.history['f1_score'])
    return  dnnMetrics    
```
